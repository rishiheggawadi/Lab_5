```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
import time
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup
import requests
import html5lib
import csv
import pandas as pd
```

```python
# Follow the 3rd Selenium tutorial on your own. 
# Submit the correct code with comments explaining what each element is doing (5pts)

PATH = Service("/Users/rishiheggawadi/Documents/HARP 151/chromedriver.exe")
driver = webdriver.Chrome(service=PATH) # sets up the driver and path

site = "https://www.techwithtim.net/" # this is the site that we will be using 
driver.get(site)
link = driver.find_element(By.LINK_TEXT, "Python Programming") #finds the link associated with the text "Python Programming"
link.click() # clicks the link

try:
    element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.LINK_TEXT, "Beginner Python Tutorials")))
    element.click() #waits 10 seconds in order to locate the link text "Beginner Python Tutorials" and clicks it
    
    element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "sow-button-19310003")))
    element.click() # waits 10 seconds in order to locate ID of the element associated with the
    # "get started" button and clicks it
    
    driver.back() 
    driver.back()
    driver.back() # goes back 3 times 
    driver.forward()
    driver.forward() # goes forward 2 times


except:
    driver.quit #if there is an error, the driver quits!
```

```python
# Use selenium to search https://www.wikipedia.org/ for the word “cat” and then scrape the url links to 
# all the images on the page. Store those in a csv file. 
# (Bonus 1pt can you add the https:// part of the url to the links so they are clickable)
PATH = Service("/Users/rishiheggawadi/Documents/HARP 151/chromedriver.exe")
driver = webdriver.Chrome(service=PATH) 

site_1 = "https://www.wikipedia.org/"
driver.get(site_1)

text = driver.find_element(By.ID, 'searchInput')
text.send_keys("cat")
text.send_keys(Keys.RETURN)

# Bonus - made them clickable links
imgResults = driver.find_elements(By.XPATH,"//img[contains(@class,'image')]")
src = []
num = 1
for img in imgResults:
    src.append(img.get_attribute('src'))
for x in src:
    link = x
    print(num,". "+link+"\n")
    num += 1
```

```python
csv_file = open("wiki_cat_data.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Link"])

for x in src:
    link = x
    csv_writer.writerow([link])

csv_file.close()

pd.read_csv("wiki_cat_data.csv")
```
